```python

def iterate_sites(sites_json):
    output = []
    for sites in range(0,len(sites_json)):

        market = sites_json[sites]['market']
        country = sites_json[sites]['address']['country']
        siteId = sites_json[sites]['_id']
        siteName = sites_json[sites]['name']
        total_consumption = 0
        for month in range(0,12):
            monthly_cons = sites_json[sites]['historyByMonth'][month]['scope2']['electricityConsumption']['amount']
            total_consumption += monthly_cons
        total_consumption
        ren_gen_percent, nonren_gen_percent = get_electricity_mix(country)
        ren_cons = round(total_consumption * ren_gen_percent)
        nonren_cons = round(total_consumption * nonren_gen_percent)
        print(total_consumption,ren_cons, nonren_cons)

        output_json = json.loads(estimated_output)
        output_json['siteID'] = siteId
        output_json['siteName'] = siteName
        output_json['country'] = country
        output_json['market'] = market
        output_json['renewableConsumption']['amount'] = ren_cons
        output_json['nonRenewableConsumption']['amount'] =  nonren_cons

        output_json
        output.append(output_json)
    return output
    
output
```

```python
def get_electricity_mix(country):
    
    elec_mix = pd.read_csv('C:\\Users\\VincentRaj\\Projects\\Consumption Analysis\\Electricity_Mix.csv')
    elec_mix.dropna(inplace=True)
    elec_mix = elec_mix[elec_mix['Country'] == country]
    
    renewables = ['Solar','Bio-fuels','Hydrogen','Hydropower','Hydro','Bio','Pumped','Tidal' ,'Ocean', 'Thermal', 'Wind' ,'Geothermal', 'Biomass' ,'Biogas']
    
    for i ,undefined_source in enumerate(elec_mix['Source']):
        if any(sources in undefined_source for sources in renewables):
            elec_mix.loc[i,'Type'] = 'Renewables'
        else:
            elec_mix.loc[i,'Type'] = 'Non-Renewables'
            
    year = elec_mix[elec_mix['Country'] == country]['Year'].max()
    
    total_gen = elec_mix[(elec_mix['Year']== year) &( elec_mix['Country'] == country)]['Generation'].sum()
    ren_gen = elec_mix[(elec_mix['Year']== year) & (elec_mix['Country'] == country) & (elec_mix['Type'] == 'Renewables')]['Generation'].sum()
    nonren_gen = elec_mix[(elec_mix['Year']== year) & (elec_mix['Country'] == country) & (elec_mix['Type'] == 'Non-Renewables') ]['Generation'].sum()
    
    nonren_gen_percent = round(nonren_gen / total_gen,2)
    ren_gen_percent = round(ren_gen / total_gen,2)

    return ren_gen_percent, nonren_gen_percent
```

```python
def get_electricity_mix(country):
    
    #elec_mix = pd.read_csv('C:\\Users\\VincentRaj\\Projects\\Consumption Analysis\\Electricity_Mix.csv')
    elec_mix = pd.read_csv('C:\\Users\\VincentRaj\\Desktop\\ElectricityMix.csv')

    #elec_mix.dropna(inplace=True)
    elec_mix = elec_mix[elec_mix['Country'] == country]
    
    renewables = ['Solar','Bio-fuels','Hydrogen','Hydropower','Hydro','Bio','Pumped','Tidal' ,'Ocean', 'Thermal', 'Wind' ,'Geothermal', 'Biomass' ,'Biogas']
    
    for i ,undefined_source in enumerate(elec_mix['Source']):
        if any(sources in undefined_source for sources in renewables):
            elec_mix.loc[i,'Type'] = 'Renewables'
        else:
            elec_mix.loc[i,'Type'] = 'Non-Renewables'
            
    year = elec_mix[elec_mix['Country'] == country]['Year'].max()
    
    total_gen = elec_mix[(elec_mix['Year']== year) &( elec_mix['Country'] == country)]['Generation'].sum()
    ren_gen = elec_mix[(elec_mix['Year']== year) & (elec_mix['Country'] == country) & (elec_mix['Type'] == 'Renewables')]['Generation'].sum()
    nonren_gen = elec_mix[(elec_mix['Year']== year) & (elec_mix['Country'] == country) & (elec_mix['Type'] == 'Non-Renewables') ]['Generation'].sum()
    
    nonren_gen_percent = round(nonren_gen / total_gen,2)
    ren_gen_percent = round(ren_gen / total_gen,2)

    return ren_gen_percent, nonren_gen_percent
```
